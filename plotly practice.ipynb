```python
import plotly.express as px
```

```python
data=px.data.tips()
```

```python
data
```

## Bar Plot

```python
fig=px.bar(x='time',y='tip',data_frame=data)
fig.show()
```

```python
fig=px.bar(x='time',y='tip',data_frame=data,color='sex')
fig.show()
```

```python
fig=px.bar(data,x='day',y='total_bill',color='smoker')
fig.show()
```

```python
fig=px.bar(data,x='day',y='total_bill',color='size')
fig.show()
```

## Bar Plot using query function

```python
fi=data.query("total_bill > 15")
fig=px.bar(fi,x='sex',y='total_bill')
fig.show()
```

```python
fi=data.query("total_bill > 15 and day =='Sun'")
fig=px.bar(fi,x='sex',y='total_bill')
fig.show()
```

```python
fi=data.query("total_bill > 15 and day =='Sun'")
fig=px.bar(fi,x='day',y='total_bill',color='sex')
fig.show()
```

```python
df=px.data.tips()
condi=df[df['time']=='Dinner']
fig=px.bar(condi,x='sex',y='total_bill',hover_data=['tip','day','size'],color='smoker')
fig.show()
```

## Scatter and bubble Sort

```python
import plotly.express as px
```

```python
df=px.data.tips()
fig=px.scatter(df,x='total_bill',y='tip',color='day',size='size',hover_data=['sex','time'])
fig.show()
```
